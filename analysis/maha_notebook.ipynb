```python
import pandas as pd
import requests
df=pd.read_csv("..\\data\\raw\\forsale.csv")

import pandas as pd
from scipy.stats import pearsonr

# 1. Drop rows where target columns or price contain missing values
clean_df = df[["bathrooms", "toilets", "price"]].dropna()

# 2. Create the combined variable (bathrooms + toilets)
clean_df["combined"] = clean_df["bathrooms"] + clean_df["toilets"]

# 3. Calculate correlation and p-value for Bathrooms vs Price
corr_bath, p_bath = pearsonr(clean_df["bathrooms"], clean_df["price"])
print(f"Bathrooms vs Price: {corr_bath} | p-value: {p_bath}")

# 4. Calculate correlation and p-value for Toilets vs Price
corr_toilet, p_toilet = pearsonr(clean_df["toilets"], clean_df["price"])
print(f"Toilets vs Price: {corr_toilet} | p-value: {p_toilet}")

# 5. Calculate correlation and p-value for Combined vs Price
corr_comb, p_comb = pearsonr(clean_df["combined"], clean_df["price"])
print(f"Combined vs Price: {corr_comb} | p-value: {p_comb}")
```

```python
import pandas as pd

raw_df = pd.read_json("./data/raw/dataframe.json")

rows_to_Int64 = ["postcode", "price", "build_year", "bedroom_count", "livable_surface",                 
                 "total_surface", "garage", "terrace", "swimming_pool", "peb_category", 
                 "Preschool_distance", "Train_station_distance", "Supermarket_distance"]

for row in rows_to_Int64:
    if raw_df[row].dtype == "float64":
        raw_df[row] = raw_df[row].astype("Int64")

new_column_names = {"peb_category": "energy_consumption_kWh/m2/year",
                    "Preschool_distance": "preschool_distance_m",
                    "Train_station_distance": "train_station_distance_m",
                    "Supermarket_distance": "supermarket_distance_m",
                    "distance_nearest_city": "nearest_city_distance_km"}

clean_df = raw_df.rename(columns=new_column_names)

clean_df.to_json("./data/cleaned/clean_dataframe.json", orient="records", force_ascii=False, indent=4)
```
