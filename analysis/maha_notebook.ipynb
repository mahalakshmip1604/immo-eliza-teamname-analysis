```python
import pandas as pd
import requests
import pandas as pd
from scipy.stats import pearsonr

df=pd.read_csv("..\\data\\cleaned\\cleaned_sale_properties.csv")

non_empty_epc_df = df[df["epc"].notna()].count()

print (non_empty_epc_df)

#TO get to know most outliners
min_max = pd.DataFrame({
    'min': df.min(),
    'max': df.max()
})
with pd.option_context(
    'display.max_rows', None,
    'display.max_columns', None,
    'display.max_colwidth', None
):
    display(min_max)


# to check duplicate based on same  location , price and surface


duplicates = df.duplicated(
    subset=[ 'latitude', 'longitude','price','livable_surface','land_surface']
).sum()

print(f"Duplicate rows: {duplicates}")


duplicated_rows = df.copy()

duplicated_rows['duplicate_count'] = (
    duplicated_rows.groupby(['latitude', 'longitude','price','livable_surface','land_surface'])
    .transform('size')
)

duplicated_rows = duplicated_rows[
    duplicated_rows['duplicate_count'] > 1
].sort_values(['latitude', 'longitude','price','livable_surface','land_surface'])

duplicated_rows.to_csv(
    'duplicate_new_summary.csv',
    index=False,
    encoding='utf-8-sig'
)
display(duplicated_rows)

```
